```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread(r"D:\Opencv\adad.jpg", 0)
filter_image = cv.GaussianBlur(image,(7,7),0)

ad_image = cv.adaptiveThreshold(filter_image, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 6)

plt.figure(figsize=[15,7])
plt.subplot(331);plt.imshow(image, cmap='gray');plt.title("Original")
plt.subplot(332);plt.imshow(filter_image, cmap='gray');plt.title("Remove Noise")
plt.subplot(333);plt.imshow(ad_image, cmap='gray');plt.title("Final")
```

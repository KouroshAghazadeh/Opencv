```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
color = (255,255,255)
def draw_triangle(event, x, y, flags, param):
    global color
    point1 = (x,y)
    point2 = (x+30, y)
    point3 = (x+15, y+30)
    if event == cv.EVENT_LBUTTONDOWN:
        cv.line(img, point1, point2, color, 3)
        cv.line(img, point2, point3, color, 3)
        cv.line(img, point1, point3, color, 3)

img = np.zeros((512,512,3), np.uint8)

cv.namedWindow('image')
cv.setMouseCallback('image', draw_triangle)

while True:
    cv.imshow('image', img)
    key = cv.waitKey(1) & 0xFF
    if key == ord('r'):
        color = (0,0,255)
    elif key == ord('g'):
        color = (0,255,0)
    elif key == ord('b'):
        color = (255,0,0)
    elif key == 27 & 0xFF:
        break
cv.destroyAllWindows()
plt.imshow(img[...,::-1])
```

```python
import numpy as np

a = np.ones([2], dtype='uint8')
b = a * 100
print(b)
print(b + 155)
print(b + 160)
```

```python
m1 = np.array([[150,155], [156,157]], dtype='uint8')
m2 = np.ones([2,2], dtype='uint8') * 100

print("m1:")
print(m1)
print("m2:")
print(m2)
print("m1+m2:")
print(m1 + m2)
```

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img1 = cv.imread("D:\Opencv\class.vision-master\images\class.vision.jpg")
shape = img1.shape
img2 = cv.imread("D:\Opencv\class.vision-master\images\input.jpg")[100:100+shape[0], 100:100+shape[1],:]

output_image = cv.add(img1,img2)

plt.figure(figsize=[10,5])
plt.subplot(131);plt.imshow(img1[...,::-1]);plt.title("image1")
plt.subplot(132);plt.imshow(img2[...,::-1]);plt.title("image2")
plt.subplot(133);plt.imshow(output_image[...,::-1]);plt.title("image1+image2")
```

```python
logo_h, logo_w, _ = img2.shape
location_h = 10
location_w = 500
padded_img2 = np.zeros_like(img1)
padded_img2[location_h:location_h+logo_h, location_w:location_w+logo_w,:] = img2
plt.imshow(padded_img2[...,::-1])
```

```python
img1 = cv.imread(r"D:\Opencv\class.vision-master\images\opencv.jpg", cv.IMREAD_UNCHANGED)
img2 = cv.imread(r"D:\Opencv\class.vision-master\images\class.vision.jpg", cv.IMREAD_UNCHANGED)
output_image = cv.addWeighted(img1, 0.3, img2, 0.7, 0)

plt.figure(figsize=[10,5])
plt.subplot(131);plt.imshow(img1[...,::-1]);plt.title("image1")
plt.subplot(132);plt.imshow(img2[...,::-1]);plt.title("image2")
plt.subplot(133);plt.imshow(output_image[...,::-1]);plt.title("output_image")
```

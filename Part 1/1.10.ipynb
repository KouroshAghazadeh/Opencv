```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.zeros((512,512,3), np.uint8)
cv.putText(image, 'Hello World!', (75, 290), cv.FONT_HERSHEY_COMPLEX, 2, (100,170,0), 3)
cv.imshow("Hello World", image)
plt.imshow(image[...,::-1])
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
image = np.zeros((512,512,3), np.uint8)
text = "سلام دنیا"
cv.putText(image, text, (75,290), cv.FONT_HERSHEY_COMPLEX, 2, (100,170,0), 3)
plt.imshow(image[...,::-1])
```

```python
import arabic_reshaper
from bidi.algorithm import get_display
from PIL import Image, ImageFont, ImageDraw

image = np.zeros((512,512,3), np.uint8)
text_to_be_reshaped = "سلام دنیا"

reshaped_text = arabic_reshaper.reshape(text_to_be_reshaped)
bidi_text = get_display(reshaped_text)

text = bidi_text.encode().decode('utf-8')

pil_image = Image.fromarray(image)
font = ImageFont.truetype("arial", 100, encoding="unic")
draw = ImageDraw.Draw(pil_image)
draw.text((75,200), text, (100,175,0), font=font)

plt.imshow(pil_image)
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image = np.zeros((512,512,3), np.uint8)

image_bw = np.zeros((512,512), np.uint8)

plt.imshow(image[...,::-1])
plt.figure()
plt.imshow(image_bw, cmap='gray')
```

```python
cv.line(image, (0,0), (511,511), (255,127,0), 5)
cv.imshow("Blue Line", image)
plt.imshow(image[...,::-1])
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
image = np.zeros((512,512,3), np.uint8)
cv.rectangle(image, (100,100), (300,250), (0,0,255), 5)
cv.imshow("Rectangle", image)
plt.imshow(image[...,::-1])
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img = cv.imread("D:\Opencv\class.vision-master\images\input.jpg")
cv.rectangle(img, (280,270), (530,400), (0,255,255), 5)
plt.imshow(img[...,::-1])
```

```python
image = np.zeros((512,512,3), np.uint8)
cv.circle(image, (350,350), 100, (15,75,50), 7)
cv.imshow("Circle", image)
plt.imshow(image[...,::-1])
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
image = np.zeros((512,512,3), np.uint8)
cv.ellipse(image, (250,250), (250,100), 40, 0, 360, (255,255,255), 5)
cv.imshow("Ellipse", image)
plt.imshow(image[...,::-1])
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
image = np.zeros((512,512,3), np.uint8)

pts = np.array([[10,50],[400,50],[90,200],[50,500]], np.int32)
pts = pts.reshape((-1,1,2))
cv.polylines(image, [pts], True, (0,0,255), 3)
cv.imshow("Polygon", image)
plt.imshow(image[...,::-1])
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread("D:\Opencv\class.vision-master\images\input.jpg")
new_image = np.zeros(image.shape, image.dtype)

alpha = float(input("Enter alpha value: "))
beta = int(input("Enter beta value: "))
new_image = cv.convertScaleAbs(image, alpha=alpha, beta=beta)

plt.figure(figsize=[15,5])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original")
plt.subplot(122);plt.imshow(new_image[...,::-1]);plt.title("Output")
```

```python
new_image2 = np.zeros(image.shape, image.dtype)

alpha = float(input("Enter alpha value: "))
beta = int(input("Enter beta value: "))

for y in range(image.shape[0]):
    for x in range(image.shape[1]):
        for c in range(image.shape[2]):
            new_image2[y, x, c] = np.clip(alpha*image[y, x, c] + beta , 0, 255)
            

plt.figure(figsize=[15,5])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original")
plt.subplot(122);plt.imshow(new_image2[...,::-1]);plt.title("Output")
```

```python
gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
gray_new_image = cv.cvtColor(new_image, cv.COLOR_BGR2GRAY)
gray_new_image2 = cv.cvtColor(new_image2, cv.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=[15,5])
plt.subplot(131);plt.hist(gray_image.ravel(), 256, [0, 256]);plt.title("Original")
plt.subplot(132);plt.hist(gray_new_image.ravel(), 256, [0, 256]);plt.title("New image")
plt.subplot(133);plt.hist(gray_new_image2.ravel(), 256, [0, 256]);plt.title("New image2")
```

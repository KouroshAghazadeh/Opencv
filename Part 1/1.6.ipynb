```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread("D:\Opencv\class.vision-master\images\input.jpg")
hsv_image = cv.cvtColor(img, cv.COLOR_BGR2HSV)

cv.imshow('HSV image', hsv_image)
cv.imshow('Hue channel', hsv_image[:,:,0])
cv.imshow('Saturation channel', hsv_image[:,:,1])
cv.imshow('Value channel', hsv_image[:,:,2])

cv.waitKey()
cv.destroyAllWindows()
```

```python
plt.figure(figsize=[20,5])
plt.subplot(131);plt.imshow(hsv_image[:,:,0], cmap='gray');plt.title("Hue Channel");
plt.subplot(132);plt.imshow(hsv_image[:,:,1], cmap='gray');plt.title("Saturation Channel");
plt.subplot(133);plt.imshow(hsv_image[:,:,2], cmap='gray');plt.title("Value Channel");
```

```python
saturated_image = cv.merge([hsv_image[:,:,0], cv.add(hsv_image[:,:,1], 50), hsv_image[:,:,2]])
saturated_image = cv.cvtColor(saturated_image, cv.COLOR_HSV2BGR)
plt.figure(figsize=[12,5])
plt.subplot(121);plt.imshow(saturated_image[...,::-1]);plt.title('Saturated_image')
plt.subplot(122);plt.imshow(img[...,::-1]);plt.title("Original Image")
```

```python
brighter_image = cv.merge([hsv_image[:,:,0], hsv_image[:,:,1], cv.add(hsv_image[:,:,2], 50)])
brighter_image = cv.cvtColor(brighter_image, cv.COLOR_HSV2BGR)
plt.figure(figsize=[12,5])
plt.subplot(121);plt.imshow(brighter_image[...,::-1]);plt.title('Berighted_image')
plt.subplot(122);plt.imshow(img[...,::-1]);plt.title("Original Image")
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
original_image = cv.imread("D:\Opencv\class.vision-master\images\sudoku.jpg", 0)

noise_removed_image = cv.medianBlur(original_image, 5)

_, output = cv.threshold(original_image, 127, 255, cv.THRESH_BINARY)
output2 = cv.adaptiveThreshold(original_image, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)
output3 = cv.adaptiveThreshold(original_image, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)

_, output4 = cv.threshold(noise_removed_image, 127, 255, cv.THRESH_BINARY)
output5 = cv.adaptiveThreshold(noise_removed_image, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)
output6 = cv.adaptiveThreshold(noise_removed_image, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)

plt.figure(figsize=[15,15])
plt.subplot(331);plt.imshow(original_image, cmap='gray');plt.title("Original")
plt.subplot(332);plt.imshow(noise_removed_image, cmap='gray');plt.title("Noise removed")

plt.subplot(334);plt.imshow(output, cmap='gray');plt.title("Global Thresholding( v = 127 )")
plt.subplot(335);plt.imshow(output2, cmap='gray');plt.title("Adaptive Mean Thresholding")
plt.subplot(336);plt.imshow(output3, cmap='gray');plt.title("Adaptive Guassian Thresholding")

plt.subplot(337);plt.imshow(output4, cmap='gray');plt.title("Global Thresholding + noise removal");
plt.subplot(338);plt.imshow(output5, cmap='gray');plt.title("Adaptive Mean + noise removal");
plt.subplot(339);plt.imshow(output6, cmap='gray');plt.title("Adaptive Gaussian + noise removal");
```

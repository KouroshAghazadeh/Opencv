```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import random
```

```python
np.random.seed = 50
random_points = np.random.randint(100,400,20)
random_points = random_points.reshape(10,2)

image = np.zeros((512,512,3), 'uint8')
for point in random_points:
    cv.circle(image, (point[0], point[1]), 8, (0,255,0), -1)
plt.imshow(image[...,::-1])
```

```python
hull = cv.convexHull(random_points)
cv.drawContours(image, [hull], -1, (0,255,255), 4)
plt.imshow(image[...,::-1])
```

```python
bgr_image = cv.imread(r"D:\Opencv\class.vision-master\images\star.png")
src_gray = cv.cvtColor(bgr_image, cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(src_gray, 127, 255, 0)

contours, _ = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

hull_list = []
for i in range(len(contours)):
    hull = cv.convexHull(contours[i])
    hull_list.append(hull)
    
cv.drawContours(bgr_image, contours, -1, (0,255,0), 1)
cv.drawContours(bgr_image, hull_list, -1, (0,0,255), 1)

plt.figure(figsize=[7,7])
plt.imshow(bgr_image[...,::-1])
```

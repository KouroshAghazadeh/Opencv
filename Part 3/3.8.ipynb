```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread(r"D:\Opencv\class.vision-master\images\shapes.png", 0)
ret, thresh = cv.threshold(image, 127, 255, 0)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
```

```python
for contour in contours:
    print(contour.shape)
```

```python
bgr_image = cv.cvtColor(image, cv.COLOR_GRAY2BGR)
cv.polylines(bgr_image, [contours[0]], True, (0,0,255), 5)
plt.imshow(bgr_image[...,::-1])
```

```python
cv.polylines(bgr_image, [contours[-1]], True, (0,255,0), 5)
plt.imshow(bgr_image[...,::-1])
```

```python
all_contours = cv.drawContours(bgr_image, contours, -1, (0,255,0), 3)
plt.imshow(all_contours)
```

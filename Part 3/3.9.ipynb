```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread(r"D:\Opencv\class.vision-master\images\rectangle.png", 0)
ret, thresh = cv.threshold(image, 127, 255, 0)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_NONE)

print("CV.CHAIN_APPROX_NONE:")
print("contours[0].shape: {} so number of points for contour: {}".format(contours[0].shape, len(contours[0])))

bgr_image = cv.cvtColor(image, cv.COLOR_GRAY2BGR)
for point in contours[0]:
    cv.circle(bgr_image, (point[0][0], point[0][1]), 2, (0, 255, 0), -1)
plt.imshow(bgr_image[...,::-1]);plt.title('CV.CHAIN_APPROX_NONE')
print('-'*5)

image = cv.imread(r"D:\Opencv\class.vision-master\images\rectangle.png", 0)
ret, thresh = cv.threshold(image, 127, 255, 0)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

print("CV.CHAIN_APPROX_SIMPLE:")
print("contours[0].shape: {} so number of points for contour: {}".format(contours[0].shape, len(contours[0])))

bgr_image = cv.cvtColor(image, cv.COLOR_GRAY2BGR)
for point in contours[0]:
    cv.circle(bgr_image, (point[0][0], point[0][1]), 2, (0, 255, 0), -1)
plt.figure();plt.imshow(bgr_image[...,::-1]);plt.title('CV.CHAIN_APPROX_SIMPLE')
print('-'*5)
```

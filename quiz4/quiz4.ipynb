```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("D:\Opencv\Weeki_Wachee_spring_10079u.jpg", cv.IMREAD_UNCHANGED)
linear_image = cv.convertScaleAbs(img, alpha=1.7, beta=-10)

lut = np.empty((1,256), 'uint8')
for i in range(256):
    lut[0,i] = np.clip(pow(i / 255.0, 0.6) * 255.0, 0, 255)
res = cv.LUT(img, lut)

plt.figure(figsize=[15,4])
plt.subplot(141);plt.imshow(img[...,::-1]);plt.title("ORG")
plt.subplot(142);plt.imshow(linear_image[...,::-1]);plt.title("Linear")
plt.subplot(143);plt.imshow(img[...,::-1]);plt.title("ORG")
plt.subplot(144);plt.imshow(res[...,::-1]);plt.title("Gamma")
```

```python
cv.imwrite("Linear.jpg", linear_image)
cv.imwrite("Gamma.jpg", res)
```

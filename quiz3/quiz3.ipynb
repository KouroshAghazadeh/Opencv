```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
image = np.zeros((512,512,3), np.uint8)
cv.rectangle(image, (250, 380), (100, 180), (255, 0, 0), 3)
plt.imshow(image[...,::-1])
```

```python
points = np.array([[180, 90], [250, 180], [100,180]], np.int32)
points = points.reshape((-1,1,2))
cv.polylines(image,[points], True, (255,0,0), 3)
plt.imshow(image[...,::-1])
```

```python
cv.imwrite("House.jpg", image)
```

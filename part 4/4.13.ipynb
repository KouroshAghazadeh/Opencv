```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
image = cv.imread(r"D:\Opencv\class.vision-master\images\xray.jpg", 0)
result = cv.equalizeHist(image)

plt.figure(figsize=[8,4])
plt.subplot(121);plt.imshow(image, cmap='gray');plt.title("Original")
plt.subplot(122);plt.imshow(result, cmap='gray');plt.title("equalized")
```

```python
plt.figure(figsize=[12,7])
plt.subplot(221);plt.imshow(image, cmap='gray');plt.title("Original")
plt.subplot(222);plt.hist(image.ravel(), 256);

plt.subplot(223);plt.imshow(result, cmap='gray');plt.title("equalizer")
plt.subplot(224);plt.hist(result.ravel(), 256);
```

```python
image = cv.imread(r"D:\Opencv\class.vision-master\images\mid.png")

channels = cv.split(image)
eq_channels = []
for ch in channels:
    eq_channels.append(cv.equalizeHist(ch))
    
eq_image = cv.merge(eq_channels)

plt.figure(figsize=[12,7])
plt.subplot(221);plt.imshow(image[...,::-1]);plt.title("Original")
plt.subplot(223);plt.imshow(eq_image[...,::-1]);plt.title("Equallized")

color = ('b', 'g', 'r')

plt.subplot(222);
for i, col in enumerate(color):
    histogram = cv.calcHist([image], [i], None, [256], [0,256])
    plt.plot(histogram, color=col)
    
plt.subplot(224);
for i, col in enumerate(color):
    histogram = cv.calcHist([eq_image], [i], None, [256], [0,256])
    plt.plot(histogram, color=col)
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
list_of_numbers = np.array([5,4,5,10,3,5,6,9,7,8,7,9,1,3,7])

plt.figure(figsize=[10,3])
plt.subplot(121);plt.hist(list_of_numbers);plt.title("Histogram")
plt.subplot(122);plt.hist(list_of_numbers, 5, color='green');plt.title("Histogram with 5 bins")
```

```python
image = cv.imread("D:\Opencv\class.vision-master\images\input.jpg", 0)

plt.figure(figsize=[10,3])
plt.subplot(121);plt.hist(image.ravel(), 256, color='orange');plt.title("Gray image hist")
plt.subplot(122);plt.hist(image.ravel(), 256, histtype='step', color='red');plt.title("Gray image hist(step)")
```

```python
image = cv.imread("D:\Opencv\class.vision-master\images\input.jpg", 0)
histogram = cv.calcHist([image], [0], None, [256], [0,256])
plt.figure(figsize=[5,3])
plt.plot(histogram)
```

```python
dark_tones = cv.imread("D:\Opencv\class.vision-master\images\dark.png")
mid_tones = cv.imread("D:\Opencv\class.vision-master\images\mid.png")
light_tones = cv.imread("D:\Opencv\class.vision-master\images\light.png")

dark_tones_gray = cv.cvtColor(dark_tones, cv.COLOR_BGR2GRAY)
mid_tones_gray = cv.cvtColor(mid_tones, cv.COLOR_BGR2GRAY)
light_tones_gray = cv.cvtColor(light_tones, cv.COLOR_BGR2GRAY)

dark_tones_hist = cv.calcHist([dark_tones_gray], [0], None, [256], [0,256])
mid_tones_hist = cv.calcHist([mid_tones_gray], [0], None, [256], [0,256])
light_tones_hist = cv.calcHist([light_tones_gray], [0], None, [256], [0,256])

plt.figure(figsize=[12,7])
plt.subplot(231);plt.imshow(dark_tones_gray, cmap='gray');plt.title("dark tones image")
plt.subplot(232);plt.imshow(mid_tones_gray, cmap='gray');plt.title("mid tones image")
plt.subplot(233);plt.imshow(light_tones_gray, cmap='gray');plt.title("light tones image")
plt.subplot(234);plt.plot(dark_tones_hist);plt.title("dark tones hist")
plt.subplot(235);plt.plot(mid_tones_hist);plt.title("mid tones hist")
plt.subplot(236);plt.plot(light_tones_hist);plt.title("light tones hist")
```

```python
plt.figure(figsize=[12,7])
plt.subplot(231);plt.imshow(dark_tones[...,::-1]);plt.title("dark tones image")
plt.subplot(232);plt.imshow(mid_tones[...,::-1]);plt.title("mid tones image")
plt.subplot(233);plt.imshow(light_tones[...,::-1]);plt.title("light tones image")

color = ('b', 'g', 'r')

plt.subplot(234);
for i, col in enumerate(color):
    histogram = cv.calcHist([dark_tones], [i], None, [256], [0,256])
    plt.plot(histogram, color=col)
    
plt.subplot(235);
for i, col in enumerate(color):
    histogram = cv.calcHist([mid_tones], [i], None, [256], [0,256])
    plt.plot(histogram, color=col)
    
plt.subplot(236);
for i, col in enumerate(color):
    histogram = cv.calcHist([light_tones], [i], None, [256], [0,256])
    plt.plot(histogram, color=col)
    
```

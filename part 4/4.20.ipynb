```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
image = cv.imread(r"D:\Opencv\class.vision-master\images\little_girl.jpg")

src = cv.GaussianBlur(image, (3,3), 0)
gray = cv.cvtColor(src, cv.COLOR_BGR2GRAY)

dst = cv.Laplacian(src, cv.CV_16S, ksize=3)

abs_dst = cv.convertScaleAbs(dst)

plt.figure(figsize=[12,7])
plt.subplot(121);plt.imshow(src[...,::-1]);plt.title("Original")
plt.subplot(122);plt.imshow(abs_dst,cmap='gray');plt.title("Laplacian")
```

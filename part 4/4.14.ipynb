```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
image = cv.imread(r"D:\Opencv\class.vision-master\images\statue.png", 0)
result = cv.equalizeHist(image)

plt.figure(figsize=[12,4])
plt.subplot(121);plt.imshow(image, cmap='gray');plt.title("Original")
plt.subplot(122);plt.imshow(result, cmap='gray');plt.title("After global eq_hist")
```

```python
image = cv.imread(r"D:\Opencv\class.vision-master\images\statue.png", 0)

clahe = cv.createCLAHE(2.0, tileGridSize = (8,8))
result = clahe.apply(image)

plt.figure(figsize=[12,4])
plt.subplot(121);plt.imshow(image, cmap='gray');plt.title("Original")
plt.subplot(122);plt.imshow(result, cmap='gray');plt.title("After...")
```

```python
image = cv.imread(r"D:\Opencv\class.vision-master\images\statue.png")

channels = cv.split(image)

clahe = cv.createCLAHE(2.0, (8,8))

eq_channel = []
for ch in channels:
    eq_channel.append(clahe.apply(ch))
eq_clahe_image = cv.merge(eq_channel)
    
plt.figure(figsize=[12,8])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original")
plt.subplot(122);plt.hist(image.ravel(), 256);

plt.subplot(221);plt.imshow(eq_clahe_image[...,::-1]);plt.title("After...")
plt.subplot(222);plt.hist(eq_clahe_image.ravel(), 256);
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread(r"D:\Opencv\class.vision-master\images\noisy-cow.png")
kernel = np.ones((3, 3), np.float32) / 9

blurred = cv.filter2D(image, -1, kernel)

kernel_7x7 = np.ones((7, 7), np.float32) / 49

blurred2 = cv.filter2D(image, -1, kernel_7x7)

plt.figure(figsize=[15,5])
plt.subplot(131);plt.imshow(image[...,::-1]);plt.title("Original")
plt.subplot(132);plt.imshow(blurred[...,::-1]);plt.title("blurred 3x3")
plt.subplot(133);plt.imshow(blurred2[...,::-1]);plt.title("blurred 7x7")
```

```python
blur = cv.blur(image, (7,7))

guassian = cv.GaussianBlur(image, (7,7), 0)
median = cv.medianBlur(image, 5)
bilateral = cv.bilateralFilter(image, 9, 75, 75)

plt.figure(figsize=[15,10])
plt.subplot(231);plt.imshow(image[...,::-1]);plt.title("Original")
plt.subplot(232);plt.imshow(blur[...,::-1]);plt.title("after uniform filter")
plt.subplot(233);plt.imshow(guassian[...,::-1]);plt.title("after non uniform guass")
plt.subplot(234);plt.imshow(median[...,::-1]);plt.title("After median filter")
plt.subplot(235);plt.imshow(bilateral[...,::-1]);plt.title("After bil filter")
```

```python
guass1 = cv.GaussianBlur(image, (11,11), 2)
guass2 = cv.GaussianBlur(image, (11,11), 3)
guass3 = cv.GaussianBlur(image, (11,11), 5)

plt.figure(figsize=[10,10])
plt.subplot(231);plt.imshow(image[...,::-1]);plt.title("Original")
plt.subplot(232);plt.imshow(guass1[...,::-1]);plt.title("1")
plt.subplot(233);plt.imshow(guass2[...,::-1]);plt.title("2")
plt.subplot(234);plt.imshow(guass3[...,::-1]);plt.title("3")
```

```python
dst = cv.fastNlMeansDenoisingColored(image, None, 6, 6, 7, 21)

plt.figure(figsize=[10,10])
plt.subplot(231);plt.imshow(image[...,::-1]);plt.title("Original")
plt.subplot(232);plt.imshow(dst[...,::-1]);plt.title("result")
```

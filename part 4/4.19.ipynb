```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread(r"D:\Opencv\class.vision-master\images\little_girl.jpg",0)

sobel_x = cv.Sobel(image, cv.CV_64F, 1, 0, ksize=5)
sobel_y = cv.Sobel(image, cv.CV_64F, 0, 1, ksize=5)

sobel = cv.add(sobel_x,sobel_y)

plt.figure(figsize=[12,7])
plt.subplot(232);plt.imshow(image,cmap='gray');plt.title("Original")
plt.subplot(234);plt.imshow(sobel_x,cmap='gray');plt.title("sobel_x")
plt.subplot(235);plt.imshow(sobel_y,cmap='gray');plt.title("sobel_y")
plt.subplot(236);plt.imshow(sobel,cmap='gray');plt.title("Final_Sobel")
```

```python
image = cv.imread(r"D:\Opencv\class.vision-master\images\little_girl.jpg")
src = cv.GaussianBlur(image,(3,3),0)
gray = cv.cvtColor(src, cv.COLOR_BGR2GRAY)


sobel_x = np.abs(cv.Sobel(image, cv.CV_16S, 1, 0, ksize=3))
sobel_y = np.abs(cv.Sobel(image, cv.CV_16S, 0, 1, ksize=3))

abs_sobelx = cv.convertScaleAbs(sobel_x)
abs_sobely = cv.convertScaleAbs(sobel_y)

sobel = cv.addWeighted(abs_sobelx,0.5, abs_sobely,0.5, 0)

plt.figure(figsize=[12,7])
plt.subplot(232);plt.imshow(src[...,::-1],cmap='gray');plt.title("Original")
plt.subplot(234);plt.imshow(abs_sobelx,cmap='gray');plt.title("sobel_x")
plt.subplot(235);plt.imshow(abs_sobely,cmap='gray');plt.title("sobel_y")
plt.subplot(236);plt.imshow(sobel,cmap='gray');plt.title("Final_Sobel")
```

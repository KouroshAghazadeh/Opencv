```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
image = cv.imread(r"D:\Opencv\class.vision-master\images\felfel-dolme.jpg")

lower = np.array([20,50,50])
upper = np.array([35,255,255])

hsv_image = cv.cvtColor(image, cv.COLOR_BGR2HSV)

mask = cv.inRange(hsv_image, lower, upper)

plt.figure(figsize=[12,6])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("image")
plt.subplot(122);plt.imshow(mask, cmap='gray');plt.title("mask")
```

```python
image = cv.imread(r"D:\Opencv\class.vision-master\images\felfel-dolme.jpg")
hsv_image = cv.cvtColor(image, cv.COLOR_BGR2HSV)

lower_red = np.array([0, 50, 50])
upper_red = np.array([10, 255, 255])
mask0 = cv.inRange(hsv_image, lower_red, upper_red)

lower_red = np.array([160, 50, 50])
upper_red = np.array([179, 255, 255])
mask1 = cv.inRange(hsv_image, lower_red, upper_red)

mask = mask0 + mask1

plt.figure(figsize=[17,7])
plt.subplot(231);plt.imshow(image[...,::-1]);plt.title("image")
plt.subplot(232);plt.imshow(mask0, cmap='gray');plt.title('lower_mask')
plt.subplot(233);plt.imshow(mask1, cmap='gray');plt.title('upper_mask')
plt.subplot(234);plt.imshow(mask, cmap='gray');plt.title('final mask')
```

```python
image = cv.imread(r"D:\Opencv\class.vision-master\images\felfel-dolme.jpg")
hsv_image = cv.cvtColor(image, cv.COLOR_BGR2HSV)

lower_red = np.array([0, 50, 50])
upper_red = np.array([10, 255, 255])
mask0 = cv.inRange(hsv_image, lower_red, upper_red)

lower_red = np.array([160, 50, 50])
upper_red = np.array([179, 255, 255])
mask1 = cv.inRange(hsv_image, lower_red, upper_red)

mask = mask0 + mask1
mask_3channel = cv.merge([mask] * 3)
result = cv.bitwise_and(image, mask_3channel)

plt.figure(figsize=[17,7])
plt.subplot(131);plt.imshow(image[...,::-1]);plt.title("image")
plt.subplot(132);plt.imshow(mask_3channel, cmap='gray');plt.title("mask with 3 channel")
plt.subplot(133);plt.imshow(result[...,::-1]);plt.title("Result")
```

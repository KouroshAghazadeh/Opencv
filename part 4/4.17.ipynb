```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
image = cv.imread(r"D:\Opencv\class.vision-master\images\plate.jpg")

kernel1 =       np.array([[0,0,0],
                          [0,2,0],
                          [0,0,0]])
kernel2 =  1/9 * np.array([[1,1,1],[1,1,1],[1,1,1]])

final_kernel = kernel1-kernel2

sharped = cv.filter2D(image, -1, final_kernel)

plt.figure(figsize=[12,7])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original")
plt.subplot(122);plt.imshow(sharped[...,::-1]);plt.title("Sharped")
```

```python
kernel_sharpening = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
sharpened = cv.filter2D(image,-1,kernel_sharpening)

plt.figure(figsize=[12,7])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original")
plt.subplot(122);plt.imshow(sharpened[...,::-1]);plt.title("Sharpened")
```

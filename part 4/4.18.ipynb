```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def print_min_max(img):
    print("Min:{}, Max:{}, dtype:{}".format(np.min(img),np.max(img), img.dtype))
```

```python
image = cv.imread(r"D:\Opencv\class.vision-master\images\gradient.png",0)

kernel = np.array([[0,0,0],[-1/2,0,1/2],[0,0,0]])

edges = cv.filter2D(image, -1, kernel)

plt.figure(figsize=[12,7])
plt.subplot(121);plt.imshow(image,cmap='gray');plt.title("Original")
plt.subplot(122);plt.imshow(edges,cmap='gray');plt.title("Edges")

print_min_max(image)
print_min_max(edges)
```

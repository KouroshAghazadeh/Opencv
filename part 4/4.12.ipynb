```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
hue_spectrum = np.linspace(0, 179, dtype='uint', num=180)

img = np.array([hue_spectrum] * 50)
saturation_channel = np.ones_like(img, dtype='uint8') * 255
value_channel = np.ones_like(img, dtype='uint8') * 255

hsv = cv.merge([img, saturation_channel, value_channel])
final_bgr_spectrum = cv.cvtColor(hsv, cv.COLOR_HSV2BGR)
```

```python
def plot_hsv_hist(image_path):
    image = cv.imread(image_path)
    hsv = cv.cvtColor(image, cv.COLOR_BGR2HSV)
    hue_hist = cv.calcHist([hsv], [0], None, [180], [0,180])
    
    plt.figure(figsize=[10,3])
    plt.subplot(121);plt.imshow(image[...,::-1]);plt.xticks([]);plt.yticks([]);
    plt.subplot(222);plt.imshow(hue_hist, color = 'red');plt.xticks([]);plt.yticks([]);
    plt.subplot(224);plt.imshow(final_bgr_spectrum[...,::-1]);plt.xticks([]);plt.yticks([]);
    
```

```python
plot_hsv_hist(r"D:\Opencv\class.vision-master\images\felfel-dolme.jpg")
```
